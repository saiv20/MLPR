```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import numpy as np
from scipy.stats import multivariate_normal  
np.random.seed(1)
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mu_0 = np.array([-1,1,-1,1])
print(mu_0)
sigma_0 = np.array([[2,-0.5,0.3,0],[-0.5,1,-0.5,0],[0.3,-0.5,1,0],[0,0,0,2]])
print(sigma_0)
```

```python
mu_1 = np.array([1,1,1,1])
print(mu_1)
sigma_1 = np.array([[1,0.3,-0.2,0],[0.3,2,0.3,0],[-0.2,0.3,1,0],[0,0,0,3]])
print(sigma_1)
```

```python
nsamples = 10000
samples = []
labels = []
Py = [0.7,0.3]
size_0 = 0
size_1 = 0

for i in range(nsamples):
    if np.random.rand() < Py[0]:
        size_0 = size_0 + 1
    else:
        size_1 = size_1 + 1

samples_0 = multivariate_normal.rvs(mean=mu_0,cov=sigma_0,size=size_0)
samples_0 = np.hstack((samples_0, np.zeros((size_0, 1)))) #adding 0 to the last column for class label
print(samples_0)

samples_1 = multivariate_normal.rvs(mean=mu_1,cov=sigma_1,size=size_1)
samples_1 = np.hstack((samples_1, np.ones((size_1, 1)))) #adding 0 to the last column for class label

print("--------------------------")
print(samples_1)
print(len(samples_0),len(samples_1))
```

```python
samples_0 = pd.DataFrame(samples_0)
#print(samples_0)
samples_1 = pd.DataFrame(samples_1)
#print(samples_1)
samples = pd.concat([samples_0,samples_1],ignore_index=True)
print(samples)
#samples.to_csv('trial_samples_2.csv')
```

```python
final_samples = pd.read_csv('trial_samples_2.csv',index_col=0).to_numpy()
discr = []
for i in range(len(final_samples)):
    samp = final_samples[i,:-1]
    disct = (multivariate_normal.pdf(samp,mean=mu_1,cov=sigma_1))/(multivariate_normal.pdf(samp,mean=mu_0,cov=sigma_0))
    discr.append(disct)
discr = np.array(discr).reshape(-1,1)
#print(discr.shape)
#print(final_samples.shape)
samples_disc = np.concatenate([final_samples,discr],axis=1)
print(final_samples)
```

```python
samples_disc = samples_disc[np.argsort(samples_disc[:, -1])]
dis_0 = samples_disc[samples_disc[:, -2] == 0][:, -1]  # Discriminants for class 0
dis_1 = samples_disc[samples_disc[:, -2] == 1][:, -1]  # Discriminants for class 1
```

```python
# Initialize DataFrame
df = pd.DataFrame(columns=['False Positive', 'True Positive', 'Gamma', 'Probability Error'])

# Iterate over samples to calculate false positive rate, true positive rate, and probability error
for discriminant in samples_disc[:,-1]:
    false_positive = np.sum(dis_0 >= discriminant) / len(dis_0)
    true_positive = np.sum(dis_1 >= discriminant) / len(dis_1)
    p_err = false_positive * 0.7 + (1 - true_positive) * 0.3
    d = {'False Positive': false_positive, 'True Positive': true_positive, 'Gamma': discriminant, 'Probability Error': p_err}
    df = df.append(d, ignore_index=True)

# Sort DataFrame based on probability error
df = df.sort_values('Probability Error')

# Get info of minimum experimental probability error
exp_min = df.iloc[0]
print('Experimental Minimum Error Info:\n')
print(exp_min)

# Calculate theoretical error
thy_gamma = 0.7 / 0.3
thy_lambdas = [np.sum(dis_0 >= thy_gamma) / len(dis_0), np.sum(dis_1 >= thy_gamma) / len(dis_1)]
thy_p_err = thy_lambdas[0] * 0.7 + (1 - thy_lambdas[1]) * 0.3
thy_min = {'False Positive': thy_lambdas[0], 'True Positive': thy_lambdas[1], 'Gamma': thy_gamma, 'Probability Error': thy_p_err}
print('Theoretical Minimum Error Info:\n')
print(thy_min)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(df['False Positive'], df['True Positive'], 'bo', markersize=4, label='ROC curve')
# Plot experimental minimum
plt.plot(exp_min['False Positive'], exp_min['True Positive'], 'ro', label='Emperical Minimum', markersize=10)
# Plot theoretical minimum
plt.plot(thy_min['False Positive'], thy_min['True Positive'], 'go', label='Theoretical Minimum', markersize=10)
plt.legend(title='Minimum Error Probabilities', loc='lower right')
plt.xlabel('Probability of False Positive')
plt.ylabel('Probability of True Positive')
plt.title('ROC Curve')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```
